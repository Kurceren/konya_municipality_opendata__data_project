```python
import pandas as pd
```

```python
df=pd.read_csv("86_202107_meshururunler.csv",sep=";")
```

```python
df=df.dropna()
```

```python
df
```

```python
df["meshur_urunler_1"]=df["meshur_urunler"]
df["meshur_urunler_2"]=df["meshur_urunler"]
df["meshur_urunler_3"]=df["meshur_urunler"]
df["meshur_urunler_4"]=df["meshur_urunler"]
df["meshur_urunler_5"]=df["meshur_urunler"]
df["meshur_urunler_6"]=df["meshur_urunler"]
df["meshur_urunler_7"]=df["meshur_urunler"]
```

```python
for x in df["meshur_urunler"].index:
    df["meshur_urunler"][x]=df["meshur_urunler"][x].replace(" ","")
    y=df["meshur_urunler"][x].split(",")
    if len(y)==1:
        df["meshur_urunler_1"][x]=y[0]
        df["meshur_urunler_2"][x]="YOK"
        df["meshur_urunler_3"][x]="YOK"
        df["meshur_urunler_4"][x]="YOK"
        df["meshur_urunler_5"][x]="YOK"
        df["meshur_urunler_6"][x]="YOK"
        df["meshur_urunler_7"][x]="YOK"
        

    if len(y)==2:
        df["meshur_urunler_1"][x]=y[0]
        df["meshur_urunler_2"][x]=y[1]
        df["meshur_urunler_3"][x]="YOK"
        df["meshur_urunler_4"][x]="YOK"
        df["meshur_urunler_5"][x]="YOK"
        df["meshur_urunler_6"][x]="YOK"
        df["meshur_urunler_7"][x]="YOK"
        

    if len(y)==3:
        df["meshur_urunler_1"][x]=y[0]
        df["meshur_urunler_2"][x]=y[1]
        df["meshur_urunler_3"][x]=y[2]
        df["meshur_urunler_4"][x]="YOK"
        df["meshur_urunler_5"][x]="YOK"
        df["meshur_urunler_6"][x]="YOK"
        df["meshur_urunler_7"][x]="YOK"
        

    if len(y)==4:
        df["meshur_urunler_1"][x]=y[0]
        df["meshur_urunler_2"][x]=y[1]
        df["meshur_urunler_3"][x]=y[2]
        df["meshur_urunler_4"][x]=y[3]
        df["meshur_urunler_5"][x]="YOK"
        df["meshur_urunler_6"][x]="YOK"
        df["meshur_urunler_7"][x]="YOK"
        

    if len(y)==5:
        df["meshur_urunler_1"][x]=y[0]
        df["meshur_urunler_2"][x]=y[1]
        df["meshur_urunler_3"][x]=y[2]
        df["meshur_urunler_4"][x]=y[3]
        df["meshur_urunler_5"][x]=y[4]
        df["meshur_urunler_6"][x]="YOK"
        df["meshur_urunler_7"][x]="YOK"

    if len(y)==6:
        df["meshur_urunler_1"][x]=y[0]
        df["meshur_urunler_2"][x]=y[1]
        df["meshur_urunler_3"][x]=y[2]
        df["meshur_urunler_4"][x]=y[3]
        df["meshur_urunler_5"][x]=y[4]
        df["meshur_urunler_6"][x]=y[5]
        df["meshur_urunler_7"][x]="YOK"
        

    if len(y)==7:
        df["meshur_urunler_1"][x]=y[0]
        df["meshur_urunler_2"][x]=y[1]
        df["meshur_urunler_3"][x]=y[2]
        df["meshur_urunler_4"][x]=y[3]
        df["meshur_urunler_5"][x]=y[4]
        df["meshur_urunler_6"][x]=y[5]
        df["meshur_urunler_7"][x]=y[6]
        
```
