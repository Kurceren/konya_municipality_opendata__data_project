```python
df.info()
```

```python
df["enlem"]=df["enlem"].astype("string")
```

```python
df["boylam"]=df["boylam"].astype("string")
```

```python
for a in df["boylam"].index:
    df["boylam"][a]=df["boylam"][a].replace(",",".")
```

```python
df["aylık/yıllık_yaralanıcı_sayısı"]= df["aylık/yıllık_yaralanıcı_sayısı"].astype("string")
```

```python
aylık=[None]*16
for aa in df["aylık/yıllık_yaralanıcı_sayısı"].index:
    df["aylık/yıllık_yaralanıcı_sayısı"][aa]=df["aylık/yıllık_yaralanıcı_sayısı"][aa].replace(".","")
```

```python
df
```

```python
df[['aylik_kullanici_sayisi', 'yillik_kullanici_sayisi']] = df['aylık/yıllık_yaralanıcı_sayısı'].str.split('/', expand=True)
```

```python
df
```

```python
df["Çay vb. içecekler"]=df["Çay vb. içecekler"].astype("string")
```

```python
for bb in df["Çay vb. içecekler"].index:
    if df["Çay vb. içecekler"][bb]=="Çay vb. içecekler":
        df["Çay vb. içecekler"][bb]="Evet"
```

```python
for bb in df["Emeklilerin_vakit_geçirebileceği_bir_tesis"].index:
    if df["Emeklilerin_vakit_geçirebileceği_bir_tesis"][bb]=="Emeklilerin vakit geçirebileceği bir tesis":
        df["Emeklilerin_vakit_geçirebileceği_bir_tesis"][bb]="Evet"
```

```python
for bb in df["Televizyon"].index:
    if df["Televizyon"][bb]=="Televizyon":
        df["Televizyon"][bb]="Evet"
```
