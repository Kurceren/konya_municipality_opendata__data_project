```python
import pandas as pd
```

```python
df= pd.read_csv("41_202106_aykome2019-1.csv",sep=";")
```

```python
df.info()
```

```python
df["cinsi"]=df["cinsi"].astype("string")
df["deger"]=df["deger"].astype("string")
df["br.fiyati"]=df["br.fiyati"].astype("string")
```

```python
df
```

```python
df["adet"]=df["br.fiyati"]
df["metre(m)"]=df["br.fiyati"]
df["metre_kare(m2)"]=df["br.fiyati"]
df["metre_kup(m3)"]=df["br.fiyati"]
df["ucretsiz"]=df["br.fiyati"]
```

```python
df.fillna("1",inplace=True)
```

```python
for x in df["deger"].index:
    if df["deger"][x]=="mt":
        df["metre(m)"][x]="+"
    else:
        df["metre(m)"][x]="0"
```

```python
for x in df["deger"].index:
    if df["deger"][x]=="m2":
        df["metre_kare(m2)"][x]="+"
    else:
        df["metre_kare(m2)"][x]="0" 
```

```python
for x in df["deger"].index:
    if df["deger"][x]=="m3":
        df["metre_kup(m3)"][x]="+"
    else:
        df["metre_kup(m3)"][x]="0"
```

```python
for x in df["deger"].index:
    if df["br.fiyati"][x]=="ucretsiz":
        df["ucretsiz"][x]="+"
    else:
        df["ucretsiz"][x]="0"
```

```python
for x in df["deger"].index:
    if (df["deger"][x]).lower()=="adet":
        df["adet"][x]="+"
    else:
        df["adet"][x]="0"
```
