```python
import pandas as pd
```

```python
df=pd.read_csv("95_202106_endemikbitkiler.csv",sep=";")
```

```python
df
```

```python
df=df.dropna()
```

```python
df
```

```python
df["endemik_(sadece_bu_bolgede_yetisen)_bitkiler"]=df["endemik_(sadece_bu_bolgede_yetisen)_bitkiler"].astype("string")
```

```python
df["endemik_bitki_1"]=df["endemik_(sadece_bu_bolgede_yetisen)_bitkiler"]
df["endemik_bitki_2"]=df["endemik_(sadece_bu_bolgede_yetisen)_bitkiler"]
df["endemik_bitki_3"]=df["endemik_(sadece_bu_bolgede_yetisen)_bitkiler"]
```

```python
for x in df["endemik_(sadece_bu_bolgede_yetisen)_bitkiler"].index:
    y=df["endemik_(sadece_bu_bolgede_yetisen)_bitkiler"][x].split(",")
    if len(y)==1:
        df["endemik_bitki_1"][x]=y[0]
        df["endemik_bitki_2"][x]="YOK"
        df["endemik_bitki_3"][x]="YOK"
        

    if len(y)==2:
        df["endemik_bitki_1"][x]=y[0]
        df["endemik_bitki_2"][x]=y[1]
        df["endemik_bitki_3"][x]="YOK"
        

    if len(y)==3:
        df["endemik_bitki_1"][x]=y[0]
        df["endemik_bitki_2"][x]=y[1]
        df["endemik_bitki_3"][x]=y[2]
        
```
