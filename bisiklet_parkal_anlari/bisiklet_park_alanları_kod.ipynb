```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("153_202203_bisikletparkalanlari.csv",sep=";")
```

```python
df.head(100)
```

```python
df.info()
```

```python
df["wkt"]=df["wkt"].astype("string")
```

```python
for a in df["wkt"].index:
    df["wkt"][a]=df["wkt"][a].replace("POINT Z ","")
```

```python
df
```

```python
df["boylam"]=df["wkt"]
df["enlem"]=df["wkt"]
```

```python
df
```

```python
for a in df["wkt"].index:
    y=df["wkt"][a].split(" ")
    df["enlem"][a]=y[1]
    df["boylam"][a]=y[0]
    
```

```python
for a in df["boylam"].index:
    df["boylam"][a]=df["boylam"][a].replace("(","")
```

```python
df
```

```python
df=df.drop(columns="wkt")
```

```python
df
```

```python
df["name"]=df["name"].astype("string")
```

```python
df["id"]=df["name"]
```

```python
df
```

```python
for a in df["name"].index:
    aa=a+1
    df["id"][a]= str(aa)
```

```python
id = df.pop('id')
df.insert(0, 'id', id)

id = df.pop('enlem')
df.insert(2, 'enlem', id)
```

```python
df
```

```python
name_column = df.pop("name")
df.insert(1, 'bisiklet_park_alanları_isimleri', name_column)
```

```python
df
```

```python
df.to_csv('bisiklet_park_alanları_csv.csv', index=False)
df.to_excel('bisiklet_park_alanları_excel.xlsx', index=False)
```
