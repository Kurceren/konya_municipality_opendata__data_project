```python
import pandas as pd
```

```python
df19= pd.read_csv("44_202106_toplamyuklemeler2019 (1).csv",sep=";",)
df20= pd.read_csv("44_202106_toplamyuklemeler2020 (1).csv",sep=";",)
df21= pd.read_csv("44_202106_toplamyuklemeler2021 (2).csv",sep=";",)
```

```python
df19
```

```python
df19["tarih"]= pd.to_datetime(df19["tarih"])
df20["tarih"]= pd.to_datetime(df20["tarih"])
df21["tarih"]= pd.to_datetime(df21["tarih"])
```

```python
df19.info()
```

```python
df19["saat"] = pd.to_timedelta(df19["saat"] ,unit="h")
df20["saat"] = pd.to_timedelta(df20["saat"] ,unit="h")
df21["saat"] = pd.to_timedelta(df21["saat"] ,unit="h")
```

```python
df19
```

```python
df19["tarih_saat"]=df19["tarih"]+df19["saat"]
df20["tarih_saat"]=df20["tarih"]+df20["saat"]
df21["tarih_saat"]=df21["tarih"]+df21["saat"]
```

```python
df19
```

```python
x = df19.pop("tarih_saat")
df19.insert(0, 'tarih_saat', x)

x1 = df20.pop("tarih_saat")
df20.insert(0, 'tarih_saat', x1)

x2 = df21.pop("tarih_saat")
df21.insert(0, 'tarih_saat', x2)
```

```python
df19
```

```python
df19=df19.drop(columns="tarih")
df20=df20.drop(columns="tarih")
df21=df21.drop(columns="tarih")
```

```python
df19=df19.drop(columns="saat")
df20=df20.drop(columns="saat")
df21=df21.drop(columns="saat")
```

```python
df19 = df19.sort_values(by="tarih_saat")
df19 = df19.reset_index(drop=True)

df20 = df20.sort_values(by="tarih_saat")
df20 = df20.reset_index(drop=True)

df21 = df21.sort_values(by="tarih_saat")
df21 = df21.reset_index(drop=True)
```
