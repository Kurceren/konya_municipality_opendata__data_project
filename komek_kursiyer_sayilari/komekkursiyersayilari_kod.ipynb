```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("33_202106_komekkursiyersayilari.csv", sep=";")
```

```python
df
```

```python
df.info()
```

```python
df=df.dropna()
```

```python
df
```

```python
df["yil"]=df["yil"].astype("string")
```

```python
df.info()
```

```python
df["kursiyer_sayisi"]=df["kursiyer_sayisi"].astype("string")
```

```python
for a in df["kursiyer_sayisi"].index:
    df["kursiyer_sayisi"][a]=df["kursiyer_sayisi"][a].replace(".","")
```

```python
df
```

```python
df["baslangic_tarihi"]=df["kursiyer_sayisi"]
df["bitis_tarihi"]=df["kursiyer_sayisi"]
```

```python
df
```

```python
for a in df["yil"].index:
    
    df["baslangic_tarihi"][a]="01-01-"+df["yil"][a]
    df["bitis_tarihi"][a]="31-12-"+df["yil"][a]
```

```python
df
```

```python
x = df.pop("baslangic_tarihi")
df.insert(0, 'baslangic_tarihi', x)
```

```python
x = df.pop("bitis_tarihi")
df.insert(1, 'bitis_tarihi', x)
```

```python
df
```
