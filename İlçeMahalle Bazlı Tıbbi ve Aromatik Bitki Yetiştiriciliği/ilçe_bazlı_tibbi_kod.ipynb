```python
import pandas as pd
```

```python
df=pd.read_csv("98_202106_tibbiaromatikbitkiler.csv",sep=";")
```

```python
df
```

```python
df=df.dropna()
```

```python
df
```

```python
df["tibbi_ve_aromatik_bitki_yetistiriciligi"]=df["tibbi_ve_aromatik_bitki_yetistiriciligi"].astype("string")
```

```python
df["tibbi_ve_aromatik_bitki_1"]=df["tibbi_ve_aromatik_bitki_yetistiriciligi"]
df["tibbi_ve_aromatik_bitki_2"]=df["tibbi_ve_aromatik_bitki_yetistiriciligi"]
df["tibbi_ve_aromatik_bitki_3"]=df["tibbi_ve_aromatik_bitki_yetistiriciligi"]
df["tibbi_ve_aromatik_bitki_4"]=df["tibbi_ve_aromatik_bitki_yetistiriciligi"]
df["tibbi_ve_aromatik_bitki_5"]=df["tibbi_ve_aromatik_bitki_yetistiriciligi"]
df["tibbi_ve_aromatik_bitki_6"]=df["tibbi_ve_aromatik_bitki_yetistiriciligi"]
df["tibbi_ve_aromatik_bitki_7"]=df["tibbi_ve_aromatik_bitki_yetistiriciligi"]
df["tibbi_ve_aromatik_bitki_8"]=df["tibbi_ve_aromatik_bitki_yetistiriciligi"]
```

```python
for x in df["tibbi_ve_aromatik_bitki_yetistiriciligi"].index:
    df["tibbi_ve_aromatik_bitki_yetistiriciligi"][x]=df["tibbi_ve_aromatik_bitki_yetistiriciligi"][x].replace(" ","")
    y=df["tibbi_ve_aromatik_bitki_yetistiriciligi"][x].split(",")
    if len(y)==1:
        df["tibbi_ve_aromatik_bitki_1"][x]=y[0]
        df["tibbi_ve_aromatik_bitki_2"][x]="YOK"
        df["tibbi_ve_aromatik_bitki_3"][x]="YOK"
        df["tibbi_ve_aromatik_bitki_4"][x]="YOK"
        df["tibbi_ve_aromatik_bitki_5"][x]="YOK"
        df["tibbi_ve_aromatik_bitki_6"][x]="YOK"
        df["tibbi_ve_aromatik_bitki_7"][x]="YOK"
        df["tibbi_ve_aromatik_bitki_8"][x]="YOK"

    if len(y)==2:
        df["tibbi_ve_aromatik_bitki_1"][x]=y[0]
        df["tibbi_ve_aromatik_bitki_2"][x]=y[1]
        df["tibbi_ve_aromatik_bitki_3"][x]="YOK"
        df["tibbi_ve_aromatik_bitki_4"][x]="YOK"
        df["tibbi_ve_aromatik_bitki_5"][x]="YOK"
        df["tibbi_ve_aromatik_bitki_6"][x]="YOK"
        df["tibbi_ve_aromatik_bitki_7"][x]="YOK"
        df["tibbi_ve_aromatik_bitki_8"][x]="YOK"

    if len(y)==3:
        df["tibbi_ve_aromatik_bitki_1"][x]=y[0]
        df["tibbi_ve_aromatik_bitki_2"][x]=y[1]
        df["tibbi_ve_aromatik_bitki_3"][x]=y[2]
        df["tibbi_ve_aromatik_bitki_4"][x]="YOK"
        df["tibbi_ve_aromatik_bitki_5"][x]="YOK"
        df["tibbi_ve_aromatik_bitki_6"][x]="YOK"
        df["tibbi_ve_aromatik_bitki_7"][x]="YOK"
        df["tibbi_ve_aromatik_bitki_8"][x]="YOK"

    if len(y)==4:
        df["tibbi_ve_aromatik_bitki_1"][x]=y[0]
        df["tibbi_ve_aromatik_bitki_2"][x]=y[1]
        df["tibbi_ve_aromatik_bitki_3"][x]=y[2]
        df["tibbi_ve_aromatik_bitki_4"][x]=y[3]
        df["tibbi_ve_aromatik_bitki_5"][x]="YOK"
        df["tibbi_ve_aromatik_bitki_6"][x]="YOK"
        df["tibbi_ve_aromatik_bitki_7"][x]="YOK"
        df["tibbi_ve_aromatik_bitki_8"][x]="YOK"

    if len(y)==5:
        df["tibbi_ve_aromatik_bitki_1"][x]=y[0]
        df["tibbi_ve_aromatik_bitki_2"][x]=y[1]
        df["tibbi_ve_aromatik_bitki_3"][x]=y[2]
        df["tibbi_ve_aromatik_bitki_4"][x]=y[3]
        df["tibbi_ve_aromatik_bitki_5"][x]=y[4]
        df["tibbi_ve_aromatik_bitki_6"][x]="YOK"
        df["tibbi_ve_aromatik_bitki_7"][x]="YOK"
        df["tibbi_ve_aromatik_bitki_8"][x]="YOK"

    if len(y)==6:
        df["tibbi_ve_aromatik_bitki_1"][x]=y[0]
        df["tibbi_ve_aromatik_bitki_2"][x]=y[1]
        df["tibbi_ve_aromatik_bitki_3"][x]=y[2]
        df["tibbi_ve_aromatik_bitki_4"][x]=y[3]
        df["tibbi_ve_aromatik_bitki_5"][x]=y[4]
        df["tibbi_ve_aromatik_bitki_6"][x]=y[5]
        df["tibbi_ve_aromatik_bitki_7"][x]="YOK"
        df["tibbi_ve_aromatik_bitki_8"][x]="YOK"
    
    if len(y)==7:
        df["tibbi_ve_aromatik_bitki_1"][x]=y[0]
        df["tibbi_ve_aromatik_bitki_2"][x]=y[1]
        df["tibbi_ve_aromatik_bitki_3"][x]=y[2]
        df["tibbi_ve_aromatik_bitki_4"][x]=y[3]
        df["tibbi_ve_aromatik_bitki_5"][x]=y[4]
        df["tibbi_ve_aromatik_bitki_6"][x]=y[5]
        df["tibbi_ve_aromatik_bitki_7"][x]=y[6]
        df["tibbi_ve_aromatik_bitki_8"][x]="YOK"

    if len(y)==8:
        df["tibbi_ve_aromatik_bitki_1"][x]=y[0]
        df["tibbi_ve_aromatik_bitki_2"][x]=y[1]
        df["tibbi_ve_aromatik_bitki_3"][x]=y[2]
        df["tibbi_ve_aromatik_bitki_4"][x]=y[3]
        df["tibbi_ve_aromatik_bitki_5"][x]=y[4]
        df["tibbi_ve_aromatik_bitki_6"][x]=y[5]
        df["tibbi_ve_aromatik_bitki_7"][x]=y[6]
        df["tibbi_ve_aromatik_bitki_8"][x]=y[7]
```
