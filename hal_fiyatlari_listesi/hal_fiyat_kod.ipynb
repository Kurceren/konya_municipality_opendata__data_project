```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("35_202106_halfiyatlari.csv",sep=";")
```

```python
df
```

```python
df.info()
```

```python
df["tarih"]=df["tarih"].astype("string")
for a in df["tarih"].index:
    
        df["tarih"][a]=df["tarih"][a].replace(".","-")
```

```python
df["tarih"] = pd.to_datetime(df["tarih"])
```

```python
df
```

```python
for a in df.columns:
    if a!="tarih":
        df[a]=df[a].astype("string").str.replace(",",".")
```

```python
df.info()
```

```python
df
```

```python
for col in df.columns:
    if col!="tarih":
        df[col] = pd.to_numeric(df[col], errors='coerce').astype('float64')
```
