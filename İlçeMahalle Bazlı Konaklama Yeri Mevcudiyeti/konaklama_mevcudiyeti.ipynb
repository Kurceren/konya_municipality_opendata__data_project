```python
import pandas as pd
```

```python
df=pd.read_csv("101_202106_konaklamayeri.csv",sep=";")
```

```python
df
```

```python
df
```

```python
df.fillna("YOK", inplace=True)
```

```python
df.drop([772], axis=0, inplace=True)
df.drop([676], axis=0, inplace=True)
df.drop([644], axis=0, inplace=True)
```

```python
df
```

```python
df["konaklama_yeri"]=df["konaklama_yeri"].astype("string")
```

```python
df["konaklama_yeri_1"]=df["konaklama_yeri"]
df["konaklama_yeri_2"]=df["konaklama_yeri"]
df["konaklama_yeri_3"]=df["konaklama_yeri"]
```

```python
df
```

```python
for x in df["konaklama_yeri"].index:
    df["konaklama_yeri"][x]=df["konaklama_yeri"][x].replace(" ","")
    y=df["konaklama_yeri"][x].split(",")
    if len(y)==1:
        df["konaklama_yeri_1"][x]=y[0]
        df["konaklama_yeri_2"][x]="YOK"
        df["konaklama_yeri_3"][x]="YOK"

    if len(y)==2:
        df["konaklama_yeri_1"][x]=y[0]
        df["konaklama_yeri_2"][x]=y[1]
        df["konaklama_yeri_3"][x]="YOK"

    if len(y)==3:
        df["konaklama_yeri_1"][x]=y[0]
        df["konaklama_yeri_2"][x]=y[1]
        df["konaklama_yeri_3"][x]=y[2]
```
